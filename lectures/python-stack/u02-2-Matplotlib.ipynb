# Matplotlib: Plotting and visualizaiton of data in Python

Matplotlib is an excellent 2D and 3D graphics library for generating scientific figures and for visualizing data. 
Some of the many advantages of this library includes:

* Easy to get started
* Support for $\LaTeX$ formatted labels and texts
* Great control of every element in a figure, including figure size and DPI. 
* High-quality output in many formats, including PNG, PDF, SVG, EPS.
* GUI for interactively exploring figures *and* support for headless generation of figure files (useful for batch jobs).

One of the of the key features of matplotlib that I would like to emphasize, and that I think makes matplotlib 
highly suitable for generating figures for scientific publications is that all aspects of the figure can be 
controlled *programmatically*. This is important for reproducibility, convenient when one needs to regenerate 
the figure with updated data or changes its appearance. 

More information can be found at the Matplotlib web page: http://matplotlib.org/

# Matplotlib and pyplot interface

The Matplotlib library has been around for some time, since almost the beginning
of Python and the scientific Python library stack.  It was originally
developed as a drop-in replacement for users of the Matlab programming
environment, so they could transition to Python and perform their
plotting and visualization tasks in familiar ways.

As a result thre is some historical cruftiness, and more than 1
API and correspondign approach you can use to create Matplotlib plots.

Current best practices highly recommends using the `pyplot` interface
structure of Matplotlib.  This is a convenient object-oriented
API approach to creating figures, and adding elements to the figures
and plots in a stateful way.  We will only use the `pyplot` api
in this class when we need to create Matplotlib based visualizations.

By convention, the `pyplot` Matplotlib api is loaded with an alias 
of `plt`, like the following:

```python
import matplotlib.pyplot as plt
```

While shortened aliases like `np` for `numpy` and `pd` for `pandas` are 
questionable for their usefulness, here the name is so long that it really does
help code readability to use the `plt` alias rather than the full namespace
name when using this API in our scripts and code..

# An Example of Building a Plot

To begin with we are going to work through creating an example plot, where
we start with basic elements, then keep adding on new features and plot elements,
till we end up with a fairly complex and well constructed visualization.
These examples were inspired from this
[Matplotlib tutorial for beginners](https://github.com/rougier/matplotlib-tutorial)

You should be familiar with NumPy and using numpy arrays before working
through this lecture notebook.

The first step is to get some data we want to visualize.  We will
simply plot the sin and cosine curves in the range from $-\pi$
to $\pi$, creating them using numpy arrays.

```python
# create the range of x values to plot
x = np.linspace(-np.pi, np.pi, 256)
```

```python
# calculate f(x) = sin(x)
s = np.sin(x)
```

```python
# calculate f(x) = cos(x)
c = np.cos(x)
```

Here `x` is a NumPy array of a sample of points within the range we want to
plot our figure on (from $-\pi$ to $\pi$.  We are going to be
plotting two curves in our figure.  So we have two sets of `y` values
or the result of applying functions to the `x` values.  Here we are going
to be plotting the $\sin()$ and $\cos()` mathematical functions.

At its most basic, to create a plot using the Matplotlib `pyplot` API
you call the `plot()` function. You need to provide 2 arrays of values for a
basic 2D plot, the `x` values, and then an array of exactly the same size of the
`y` value locations to be plotted for each `x`.  By default an instance of a
`Figure` object will be created the first time you call `plot()`.  Also by
default, a line plot will be done, with lines connecting each of the subsequent
points you plot.  So the result of plotting our $\sin()$ function is.

```python
plt.plot(x, s);
```

As an aside, the default figure size is a bit small for the figures produces.
I often change the default figure size in notebooks that have a lot of figure
plots, like the following.

```python
# set a default of 10 inches width by 8 inches heght for plots
plt.rcParams['figure.figsize'] = (10, 8)
```

```python
plt.plot(x, s);
```

If you plot the $\cos()$ you will set the plot on a different plot.

```python
plt.plot(x, c);
```

As we mentioned, the `pyplot` interface is object-oriented.  If you perform
multiple plotting actions in the same shell or script, a common `Figure`
object will be created and used to add in and change state as you request
multiple actions of your figre. 

So in short, if we want both line plots on the same plot, we should 
call `plot()` two times in the same cell context.

```python
plt.plot(x, s)
plt.plot(x, c);
```

Matplotlib has a lot of default settings that are used and can be customized.
For instance, by default the first line plotted in this style gets a blue 
color, and the net one an orange color.  Also the x and y axis of the figure will
be scaled to only show the range of values in the give x and y values of the plot.
So our x values only range from $-\pi$ to $\pi$ and thus our x axis has that range,
and the $\sin$ and $\cos$ functions will result in values that range from -1 to
1 as we can see in the plot.

In the next cell we redo the plot.  However, we have added in
functions to explicitly set most all of these properties.  We set them here
to the defaults you just saw above, but by changeing the properties
in the next cell you can effect the line colors, style, x and y range and tick
marks, etc.

```python
# create a new figure, the default is 8x6, we changed default to 10x8
# default is figures rendered at 75dpi
plt.figure(figsize=(10,8), dpi=75)

# create a new subplot of the plot.  We use a grid of 1x1, so only a single
# plot, but later we can use this to have multiple subplots in the same figure
plt.subplot(1, 1, 1)

# plot the sine function using the style color 0 C0, continuous line of width 1.5
plt.plot(x, s, color='C0', linewidth=1.5, linestyle='-')

# plot the cosine function using the style color 1 C1 color continuous line of width 1.5
plt.plot(x, c, color='C1', linewidth=1.5, linestyle='-')

# set the x limits, default is something like +- 10% of the min and max range, we
# cans simulate this using min() and max() functions
plt.xlim(x.min()*1.1, x.max()*1.1)

# set the x ticks , use the step size we saw from default.  stop at 3.1 so point 3.0 is included
plt.xticks(np.arange(-3, 3.1, 1))

# set the y limits, again we will add a +/- 10% around the plot
plt.ylim(c.min()*1.1, c.max()*1.1)

# set y ticks, use the step size we saw from default.  stop at 1.1 so point 1.0 is included
plt.yticks(np.arange(-1, 1.1, 0.25));
```

## Setting tick locations and labels

The current x ticks and y ticks are not ideal.  Especially the x tick marks do not show the interesting
values, which are fractions of $\pi$, and the ends are at $-\pi$ and $\pi$.  Locations where the $\sin$
and $\cos$ functions are at their minimum and maximum will be at $\pm\pi$ and $\pm \frac{\pi}{2}$

The y ticks are maybe a bit too crowded, and really only the min and max values, and the 0 axis are of importance
in this plot.

Lets fix the tick marks.

```python
# create a new figure, the default is 8x6, we changed default to 10x8
# default is figures rendered at 75dpi
plt.figure(figsize=(10,8), dpi=75)

# create a new subplot of the plot.  We use a grid of 1x1, so only a single
# plot, but later we can use this to have multiple subplots in the same figure
plt.subplot(1, 1, 1)

# plot the sine function using the style color 0 C0, continuous line of width 1.5
plt.plot(x, s, color='blue', linewidth=2.5, linestyle='-')

# plot the cosine function using the style color 1 C1 color continuous line of width 1.5
plt.plot(x, c, color='red', linewidth=2.5, linestyle='-')

# set the x limits, default is something like +- 10% of the min and max range, we
# cans simulate this using min() and max() functions
plt.xlim(x.min()*1.1, x.max()*1.1)

# set the x ticks, begin and end at pi, and use increments of pi/2
plt.xticks( np.arange(-np.pi, np.pi+0.1, np.pi/2.0))

# set the y limits, again we will add a +/- 10% around the plot
plt.ylim(c.min()*1.1, c.max()*1.1)

# set y ticks, lets display one -1, 0 and 1
plt.yticks(np.arange(-1, 1.1, 1));
```

Getting better, but we really want the x ticks labeled as pi, not the real valued approximation of these
tick marks.  We can make this explicit.  

Notice that Matplotlib accepts \LaTeX math markup for labels, thus we can label the x ticks with their
mathematical expressions.

We will also use math mode labels for your y ticks as well so visually they are in the same font and style here.

```python
# create a new figure, the default is 8x6, we changed default to 10x8
# default is figures rendered at 75dpi
plt.figure(figsize=(10,8), dpi=75)

# create a new subplot of the plot.  We use a grid of 1x1, so only a single
# plot, but later we can use this to have multiple subplots in the same figure
plt.subplot(1, 1, 1)

# plot the sine function using the style color 0 C0, continuous line of width 1.5
plt.plot(x, s, color='blue', linewidth=2.5, linestyle='-')

# plot the cosine function using the style color 1 C1 color continuous line of width 1.5
plt.plot(x, c, color='red', linewidth=2.5, linestyle='-')

# set the x limits, default is something like +- 10% of the min and max range, we
# cans simulate this using min() and max() functions
plt.xlim(x.min()*1.1, x.max()*1.1)

# set the x ticks, begin and end at pi, and use increments of pi/2
plt.xticks( [-np.pi,    -np.pi/2,    0,      np.pi/2,    np.pi],
            [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'+$\pi$'])

# set the y limits, again we will add a +/- 10% around the plot
plt.ylim(c.min()*1.1, c.max()*1.1)

# set y ticks, lets display one -1, 0 and 1
plt.yticks([-1,      0,      1],
           [r'$-1$', r'$0$', r'$+1$']);
```

## Moving spines

Spines are the lines connecting the axis tick marks and noting the boundaries of the data area.
For many figures we keep them like we have in the previous plot, bounding our figure.

But for a mathematical plot, mathemeticians are used to having the spines go through the 0,0
origin point. 

There were actually 4 spines in the figure, left, right, top and bottom.  We will
discard all of them and instead specify new spines at $x = 0$ and $y = 0$.

To do this next part, we need to dig a bit into the object-oriented api.
To effect the spines, we need to get ahold of the current axis objects of our figure,
which are `Axis` objects, and move them or remove them.  Actually we will just make some
of them invisible (set their color to none), and move 2 of them to where we want them.

```python
# create a new figure, the default is 8x6, we changed default to 10x8
# default is figures rendered at 75dpi
plt.figure(figsize=(10,8), dpi=75)

# create a new subplot of the plot.  We use a grid of 1x1, so only a single
# plot, but later we can use this to have multiple subplots in the same figure
ax = plt.subplot(1, 1, 1)

# change the figure spines.  Note in previous step we got the subplot axis now which are
# returned when you create a subplot.  We created the subplot explicitly before so we
# could easily get the axes objects

# now just make two of the spines invisible and don't use them
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# and move the other two spines
ax.xaxis.set_ticks_position('bottom') # tick labels below axis, this is actually where they already are
ax.spines['bottom'].set_position(('data', 0)) # moves position of bottom spine to y=0

ax.yaxis.set_ticks_position('left') # tick labels to left of axis, again this was where they were already
ax.spines['left'].set_position(('data', 0)) # moves position of left spine to x=0

# plot the sine function using the style color 0 C0, continuous line of width 1.5
plt.plot(x, s, color='blue', linewidth=2.5, linestyle='-')

# plot the cosine function using the style color 1 C1 color continuous line of width 1.5
plt.plot(x, c, color='red', linewidth=2.5, linestyle='-')

# set the x limits, default is something like +- 10% of the min and max range, we
# cans simulate this using min() and max() functions
plt.xlim(x.min()*1.1, x.max()*1.1)

# set the x ticks, begin and end at pi, and use increments of pi/2
plt.xticks( [-np.pi,    -np.pi/2,    0,      np.pi/2,    np.pi],
            [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'+$\pi$'])

# set the y limits, again we will add a +/- 10% around the plot
plt.ylim(c.min()*1.1, c.max()*1.1)

# set y ticks, lets display one -1, 0 and 1
plt.yticks([-1,      0,      1],
           [r'$-1$', r'$0$', r'$+1$']);
```

Notice that we are digging into some of the internal objects of the `pyplot` API here.  We get ahold of our
`Axis` object of our subplot, through which we have access to the figure spines.  The spines are actually
a dictionary of attributes, so we use a key like `spines['left']` to get the original left spine of the figure and
send messages to it.

## Adding a legend

Whenver you have 2 or more plots on the same figure, you should label the differnt plots involved and have
a figure legend.

Here are two plots are the sine and cosine functions.  Add a legend by labeling our lines and then
invoking the `plt.legend()` function to display the legend.

While we are at it, the origin have 2 0 labels is a bit redundant, lets remove the label from the y
axis for our origin.

```python
# create a new figure, the default is 8x6, we changed default to 10x8
# default is figures rendered at 75dpi
plt.figure(figsize=(10,8), dpi=75)

# create a new subplot of the plot.  We use a grid of 1x1, so only a single
# plot, but later we can use this to have multiple subplots in the same figure
ax = plt.subplot(1, 1, 1)

# change the figure spines.  Note in previous step we got the subplot axis now which are
# returned when you create a subplot.  We created the subplot explicitly before so we
# could easily get the axes objects

# now just make two of the spines invisible and don't use them
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# and move the other two spines
ax.xaxis.set_ticks_position('bottom') # tick labels below axis, this is actually where they already are
ax.spines['bottom'].set_position(('data', 0)) # moves position of bottom spine to y=0

ax.yaxis.set_ticks_position('left') # tick labels to left of axis, again this was where they were already
ax.spines['left'].set_position(('data', 0)) # moves position of left spine to x=0

# plot the sine function using the style color 0 C0, continuous line of width 1.5
plt.plot(x, s, color='blue', linewidth=2.5, linestyle='-', label=r'$\sin \; (x)$')

# plot the cosine function using the style color 1 C1 color continuous line of width 1.5
plt.plot(x, c, color='red', linewidth=2.5, linestyle='-', label=r'$\cos \; (x)$')

# set the x limits, default is something like +- 10% of the min and max range, we
# cans simulate this using min() and max() functions
plt.xlim(x.min()*1.1, x.max()*1.1)

# set the x ticks, begin and end at pi, and use increments of pi/2
plt.xticks( [-np.pi,    -np.pi/2,    0,      np.pi/2,    np.pi],
            [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'+$\pi$'])

# set the y limits, again we will add a +/- 10% around the plot
plt.ylim(c.min()*1.1, c.max()*1.1)

# set y ticks, lets display one -1, 0 and 1
plt.yticks([-1,      1],
           [r'$-1$', r'$+1$'])
         
# add the legend to the figure
plt.legend(loc='upper left', frameon=False);
```

## Annotate some points

Often to create a final figure for your research paper or report to submit, you want to tell a story
with your figure.  Sometimes it is best to just explicitly annotate your figure with
some text to point out details you want to draw attention to.

Here just as an example, lets say the value at $2 \pi / 3$ is special, and we want to annotate the figure 
and explicitly point out some aspect of the plot at that point for our two line plots.

We'll first add some plotting elements by drawing a marker and dotted lines from the x location to the
curves.  Then we'll use the annotate command to display some text and arrows.

```python
# create a new figure, the default is 8x6, we changed default to 10x8
# default is figures rendered at 75dpi
plt.figure(figsize=(10,8), dpi=75)

# create a new subplot of the plot.  We use a grid of 1x1, so only a single
# plot, but later we can use this to have multiple subplots in the same figure
ax = plt.subplot(1, 1, 1)

# change the figure spines.  Note in previous step we got the subplot axis now which are
# returned when you create a subplot.  We created the subplot explicitly before so we
# could easily get the axes objects

# now just make two of the spines invisible and don't use them
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# and move the other two spines
ax.xaxis.set_ticks_position('bottom') # tick labels below axis, this is actually where they already are
ax.spines['bottom'].set_position(('data', 0)) # moves position of bottom spine to y=0

ax.yaxis.set_ticks_position('left') # tick labels to left of axis, again this was where they were already
ax.spines['left'].set_position(('data', 0)) # moves position of left spine to x=0

# plot the sine function using the style color 0 C0, continuous line of width 1.5
plt.plot(x, s, color='blue', linewidth=2.5, linestyle='-', label=r'$\sin \; (x)$')

# plot the cosine function using the style color 1 C1 color continuous line of width 1.5
plt.plot(x, c, color='red', linewidth=2.5, linestyle='-', label=r'$\cos \; (x)$')

# set the x limits, default is something like +- 10% of the min and max range, we
# cans simulate this using min() and max() functions
plt.xlim(x.min()*1.1, x.max()*1.1)

# set the x ticks, begin and end at pi, and use increments of pi/2
plt.xticks( [-np.pi,    -np.pi/2,    0,      np.pi/2,    np.pi],
            [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'+$\pi$'])

# set the y limits, again we will add a +/- 10% around the plot
plt.ylim(c.min()*1.1, c.max()*1.1)

# set y ticks, lets display one -1, 0 and 1
plt.yticks([-1,      1],
           [r'$-1$', r'$+1$'])
         
# add the legend to the figure
plt.legend(loc='upper left', frameon=False)

# add figure elements to show line from point 2pi/3 to our two line plots

# first a blue dotted line and marker to the sine plot
x_pos = 2*np.pi / 3
plt.plot([x_pos, x_pos], [0, np.sin(x_pos)], color='blue', linewidth=1.5, linestyle='--') # dashed line from x axis to sin
plt.scatter([x_pos], [np.sin(x_pos)], 50, color = 'blue') # add a blue marker at position on sin 

# annotate the sine plot at this notable position
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$',
             xy=(x_pos, np.sin(x_pos)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))

# next a dotted line and marker to the cosine plot
plt.plot([x_pos, x_pos], [0, np.cos(x_pos)], color='red', linewidth=1.5, linestyle='--') # dashed line from x axis to cosin
plt.scatter([x_pos], [np.cos(x_pos)], 50, color = 'red') # add a red marker at position on cosin 

# annotate the cosine plot at this notable position
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$',
             xy=(x_pos, np.cos(x_pos)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'));
```

## Devil is in the details

This is a pretty nice figure, but there are definitely some more things that could be improved.
The tick marks are a bit small and some are obscured by the line plots.  In general the font
is a bit small, for the legend as well.  And a final figure should probably have a figure title
usually for inclusion in a presentation or a paper.

Setting the fontsize for the legend and a title can be easily accomplished in the `plt.legend()`
and by using the `plt.title()` functions.

Setting font properties for tic marks is a bit tougher.  We have to access them again through the

```python
# create a new figure, the default is 8x6, we changed default to 10x8
# default is figures rendered at 75dpi
plt.figure(figsize=(10,8), dpi=75)

# create a new subplot of the plot.  We use a grid of 1x1, so only a single
# plot, but later we can use this to have multiple subplots in the same figure
ax = plt.subplot(1, 1, 1)

# change the figure spines.  Note in previous step we got the subplot axis now which are
# returned when you create a subplot.  We created the subplot explicitly before so we
# could easily get the axes objects

# now just make two of the spines invisible and don't use them
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# and move the other two spines
ax.xaxis.set_ticks_position('bottom') # tick labels below axis, this is actually where they already are
ax.spines['bottom'].set_position(('data', 0)) # moves position of bottom spine to y=0

ax.yaxis.set_ticks_position('left') # tick labels to left of axis, again this was where they were already
ax.spines['left'].set_position(('data', 0)) # moves position of left spine to x=0

# added so that the axis elements will be on top of the plot elements, so tic marks are visible
ax.set_axisbelow(False)

# plot the sine function using the style color 0 C0, continuous line of width 1.5
plt.plot(x, s, color='blue', linewidth=2.5, linestyle='-', label=r'$\sin \; (x)$')

# plot the cosine function using the style color 1 C1 color continuous line of width 1.5
plt.plot(x, c, color='red', linewidth=2.5, linestyle='-', label=r'$\cos \; (x)$')

# set the x limits, default is something like +- 10% of the min and max range, we
# cans simulate this using min() and max() functions
plt.xlim(x.min()*1.1, x.max()*1.1)

# set the x ticks, begin and end at pi, and use increments of pi/2
plt.xticks( [-np.pi,    -np.pi/2,    0,      +np.pi/2,    +np.pi],
            [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'+$\pi$'])

# set the y limits, again we will add a +/- 10% around the plot
plt.ylim(c.min()*1.1, c.max()*1.1)

# set y ticks, lets display one -1, 0 and 1
plt.yticks([-1,      +1],
           [r'$-1$', r'$+1$'])
         
# add the legend to the figure
plt.legend(loc='upper left', frameon=False, fontsize=20)

# add figure elements to show line from point 2pi/3 to our two line plots

# first a blue dotted line and marker to the sine plot
x_pos = 2*np.pi / 3
plt.plot([x_pos, x_pos], [0, np.sin(x_pos)], color='blue', linewidth=1.5, linestyle='--') # dashed line from x axis to sin
plt.scatter([x_pos], [np.sin(x_pos)], 50, color = 'blue') # add a blue marker at position on sin 

# annotate the sine plot at this notable position
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$',
             xy=(x_pos, np.sin(x_pos)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))

# next a dotted line and marker to the cosine plot
plt.plot([x_pos, x_pos], [0, np.cos(x_pos)], color='red', linewidth=1.5, linestyle='--') # dashed line from x axis to cosin
plt.scatter([x_pos], [np.cos(x_pos)], 50, color = 'red') # add a red marker at position on cosin 

# annotate the cosine plot at this notable position
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$',
             xy=(x_pos, np.cos(x_pos)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))

# fix the tick mark font properties.  While we are at it, add a bounding box so they
# are above but a bit transparent the plot elements
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))
    
plt.title(r'Interesting Properties of Sine and Cosine Functions at $x = \frac{2\pi}{3}$', fontsize=20);
```

## Example dataset

Before we plot som stuff, lets create an example dataset.  In most of this class we will use datasets that we load from a 
file or some source for analysis.  But we can create one of our own here to demonstrate some basic plotting in Matplotlib.

We will generate random data but with some properties, like some correlations, so we will have some interesting things to show
in our plots.  

We will create a dictionary where the keys are the attributes, and each value will be a NumPy 1-D array, all of the same size.
The most recent versions of Matplotlib support plotting data sources that can access attributes by name.  This is becoming a 
common way to organize a data set.  We will look at and use Pandas `DataFrame` objects in the next video, but they are basically
organized tables of data that can be referenced by attribute names like this.

I won't give too much description of the example dataset created next here as it is mostly only made to demonstrate Matplotlib.
But we will simulate a rather large set of samples of people with various attributes like age, weight, height, and some other
properties we can use to demonstrate some basic plots with.

Sources:

- [US Census: Age and Sex distribution 2010](https://www.census.gov/prod/cen2010/briefs/c2010br-03.pdf)
- [National Health Statistics Reports, weight, height, etc.](https://www.cdc.gov/nchs/data/nhsr/nhsr122-508.pdf)

```python
# Number of subjects/samples in our dataset
m = 10000

# dataset will be organized as a dictionary, with attribute names for keys and numpy arrays all of the same size for
# values
data = {}

# create a unique id attribute, just a number from 1 to m
id = np.arange(1, m+1)
data['id'] = id

# simulate peoples ages by sampling from a distribution as described in US Census
# create a discrete category, the number is just category id, the tuple contains age begin, age end, percentage of population
age_distribution = {
# cat   beg end  %
    0: ( 0,   4,  6.5),
    1: ( 5,  17, 17.5),
    2: (18,  24,  9.9),
    3: (25,  44, 26.6),
    4: (45,  64, 26.4),
    5: (65, 100, 13.1), # total because of rounding is only 99.9, so add 0.1 to last category
} 
age_categories = list(age_distribution.keys())
age_labels = ['0-4', '5-17', '18-24', '25-44', '45-64', '65-100']
age_ratios = np.array([pct/100.0 for (begin_age, end_age, pct) in age_distribution.values()])
age_category = np.random.choice(age_categories, size=(m,), p=age_ratios)
data['age_category'] = age_category

# create actual ages, draw uniformly from the corresponding age_category.
age_ranges = [(begin_age, end_age) for (begin_age, end_age, pct) in age_distribution.values()]

def random_uniform_age(age_category):
    """Given an age category, for each category, create a randomly generate age with a uniform
    distribution in the indicated range and return it.
    
    NOTE: this function uses an external age_ranges[] array to map from the category to the
    age range.
    """
    begin_age, end_age = age_ranges[age_category]
    age = np.random.uniform(begin_age, end_age, 1)
    return age

age = np.array(list(map(random_uniform_age, age_category))).flatten()
data['age'] = age


# sex is a binary variable (in the census data).  2010 US census shows 49.2% population male and 50.8% female
male, female = 0, 1 # sex category variable
sex = np.random.choice([male, female], size=(m,), p=[0.492, 0.508])
data['sex'] = sex


# random weights by sex
# male: mean 197.9lb, standard error=2.0 (20 and over)
# female: mean 170.5lb, standard error=1.7 (20 and over)
def random_normal_weight_by_sex_age(sex, age):
    """Given a sex category, 0 for male 1 for female, and the persons age
    Generate a random weight.  If age is 20 and over we sample from a normal
    distribution by sex.  If age is less than 20 we decrease linearly with some
    random noise down to 7lb 
    """
    # constants used in this function
    n = 45047 # number of participants sampled in the data we are using to estimate distributions
    male_mu = 197.9
    male_sd = 0.2 * np.sqrt(n)
    female_mu = 170.5
    female_sd = 0.17 * np.sqrt(n)
    
    # determine a random weight depending on male or female distribution
    if sex == male:
        weight = np.random.normal(male_mu, male_sd, 1)
    else: # female
        weight = np.random.normal(female_mu, female_sd, 1)
            
    # linearly scale for children under 20 by age in range from 7.0lb to weight.
    if age < 20:
        range = weight - 7.0
        scale = age / 20
        weight = 3.5 + (range * scale)
        
    return weight

weight = np.array(list(map(random_normal_weight_by_sex_age, sex, age))).flatten()
data['weight'] = weight

# random height.  Thee is a fairly strong correlation between peoples weight and height
# so we will create a function that maps from weight to height, using sex as well,
# but also adds some random noise since the correlation is not perfect
# mean height of males is 69.1 inches standard error 0.1
# mean height of females is 63.7 inches standard error 0.1
def weight_to_height(weight, sex):
    """Given the weight and sex of an individual, create a correlated random height.
    """
    male_height_mu, male_weight_mu = 69.1, 197.9
    female_height_mu, female_weight_mu = 63.7, 170.5
    
    if sex == male:
        height = np.random.normal(male_height_mu, 5, 1)     
    else: # sex == female:
        height = np.random.normal(female_height_mu, 4, 1)
        
    # adjust by the weight ratio to add some correlation
    ratio = (weight - male_weight_mu) / male_weight_mu
    #print(ratio/10.0)
    height = height + (ratio/10.0 * height)
        
    return height

height = np.array(list(map(weight_to_height, weight, sex))).flatten()
data['height'] = height
```

## Bar chart

You can plot simple bar charts with matplotlib.  More complex options include
grouping or stacking bars.  You can have vertical or horizontal bars, etc.

Bar charts are often used to summarize information.  So for example, we might
want to have a bar chart with error bars of our population weight or height.

```python
# example of boolean indexing to separate the samples into groups
male_indexes = (sex == male)
male_weight = weight[male_indexes]

female_indexes = (sex == female)
female_weight = weight[female_indexes]

# calculate the mean and sd and put into a sequence
male_weight_mean, female_weight_mean = male_weight.mean(), female_weight.mean()
male_weight_std, female_weight_std = male_weight.std(), female_weight.std()

# convert to standard error
male_weight_se = male_weight_std / np.sqrt(len(male_weight))
female_weight_se = female_weight_std / np.sqrt(len(female_weight))
```

```python
# for readability, these are the information we are plotting
categories = [male, female]
labels=['men', 'women']
weights = [male_weight_mean, female_weight_mean]
errors = [male_weight_std, female_weight_std]

# plot the bars
rects = plt.bar(categories, weights, yerr=errors,
                alpha=0.5, error_kw=dict(lw=2, capsize=10, capthick=2, ecolor='grey'))

plt.xticks(categories, labels)
plt.ylabel('weight (lb)')
plt.xlim([-0.75, 1.75])

# add a label above each bar of the bar value
def autolabel(rects):
    ax = plt.gca()
    for rect in rects:
        height = rect.get_height()
        ax.annotate('%0.1f' % (height),
                   xy=(rect.get_x() + rect.get_width() / 2, height),
                   xytext=(-25,3), # 3 points vertical offset
                   textcoords="offset points",
                   ha='center', va='bottom')
autolabel(rects)        

plt.title('Average Population Weight by Sex (with std error bars)');
```

Can make the bars horizontal, just as an example.

```python
# # separate out height samples
male_height = height[male_indexes]
female_height = height[female_indexes]

# calculate the mean and sd and put into a sequence
male_height_mean, female_height_mean = male_height.mean(), female_height.mean()
male_height_std, female_height_std = male_height.std(), female_height.std()

# convert to standard error
male_height_se = male_height_std / np.sqrt(len(male_height))
female_height_se = female_height_std / np.sqrt(len(female_height))
```

```python
# for readability, these are the information we are plotting
categories = [male, female]
labels=['men', 'women']
heights = [male_height_mean, female_height_mean]
errors = [male_height_std, female_height_std]

plt.barh(categories, heights, xerr=errors,
         alpha=0.5, error_kw=dict(lw=2, capsize=10, capthick=2, ecolor='grey'))

plt.yticks(categories, labels)
plt.xlabel('height (in)')
plt.ylim([-0.75, 1.75])
plt.title('Average Population Height by Sex (with std error bars)');
```

You can also group bar charts together.  Here we will color code by sex, and plot weight and height
as two separate groups of bars.

```python
# for readability, these are the information we are plotting
categories = np.array([0, 1])
labels=['weight', 'height']

men = [male_weight_mean, male_height_mean]
men_errors = [male_weight_std, male_height_std]

women = [female_weight_mean, female_height_mean]
women_errors = [female_weight_std, female_height_std]


# plot the bars
width=0.35 # the width of the bars
rects_men = plt.bar(categories-width/2, men, width, yerr=men_errors, color='C0', label='men',
                    alpha=0.5, error_kw=dict(lw=2, capsize=10, capthick=2, ecolor='grey'))
rects_women = plt.bar(categories+width/2, women, width, yerr=women_errors, color='C1', label='women',
                      alpha=0.5, error_kw=dict(lw=2, capsize=10, capthick=2, ecolor='grey'))

plt.xticks(categories, labels)
plt.ylabel('weight (lb) / height (in)')
plt.xlim([-0.75, 1.75])

# add a label above each bar of the bar value
def autolabel(rects):
    ax = plt.gca()
    for rect in rects:
        height = rect.get_height()
        ax.annotate('%0.1f' % (height),
                   xy=(rect.get_x() + rect.get_width() / 2, height),
                   xytext=(-25,3), # 3 points vertical offset
                   textcoords="offset points",
                   ha='center', va='bottom')
autolabel(rects_men) 
autolabel(rects_women)

plt.legend(fontsize=16)

plt.title('Average Population Weight by Sex (with std error bars)');
```

## Line plot

Fit a line and also show filling between lines for a confidence interval.

```python
# add in color by sex
scatter = plt.scatter('weight', 'height', c='sex', data=data, alpha=0.25, cmap=plt.cm.viridis)
handles, labels = scatter.legend_elements(prop='colors')
plt.legend(handles, ['male', 'female'], loc='upper left', title='sex')
plt.xlabel('weight (lb)')
plt.ylabel('height (in)')
plt.title('Scatter Plot of Weight vs. Height by male/female');


# fit a linear model to the weight/height relationship
slope, intercept = np.polyfit(weight, height, deg=1)

# to estimate error, need weights in sorted order
x = np.sort(weight) # make a copy of the data and sort it from lowest to highest

# our linear estimate
y_est = slope * x + intercept

# estimate of the error on model
y_err = x.std() * np.sqrt(1.0/len(x) +
                          (x - x.mean())**2.0 / np.sum((x - x.mean())**2))
    
# display the line
plt.plot(x, y_est, '-')

# display the error
plt.fill_between(x, y_est-y_err, y_est+y_err, alpha=0.2);
```
